```python
import pandas as pd
data={
    "cutomer id":[1,2,3,4,5],
    "feedback":[
        "The good proudrct","Staff is good","place is clean","parking is avilable","food is testy"
    ]
}
df=pd.DataFrame(data)
print(df)
```

```python
def pre_process_text(text5):
  words=word_tokenize(text5)

  words=[w for w in words if w not in stop]

  words=[w for w in words if w not in string.punctuation]

  words=[w.lower() for w in words]

  return words
```

```python
df['clean feedback']=df['feedback'].apply(pre_process_text)
print(df)
```
