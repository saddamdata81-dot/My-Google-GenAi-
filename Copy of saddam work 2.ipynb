```python
  # Install required packages
!pip install transformers torch sentencepiece

from transformers import pipeline

# Create multiple translators
translator_en_de = pipeline("translation_en_to_de", model="Helsinki-NLP/opus-mt-en-de")
translator_en_fr = pipeline("translation_en_to_fr", model="Helsinki-NLP/opus-mt-en-fr")
translator_en_hi = pipeline("translation_en_to_hi", model="Helsinki-NLP/opus-mt-en-hi")

# Translate to different languages
text = "Hello, how are you?"

german = translator_en_de(text)[0]['translation_text']
french = translator_en_fr(text)[0]['translation_text']
hindi = translator_en_hi(text)[0]['translation_text']

print(f"German: {german}")
print(f"French: {french}")
print(f"Hindi: {hindi}")
```
