```python
!pip install -q faiss-cpu
from sentence_transformers import SentenceTransformer
import numpy as np
import faiss
```

```python
with open("/content/Food Recepies.txt",'r') as file :
  text  = [line.strip() for line in file if line.strip()]
  print(text)

model = SentenceTransformer('all-MinilM-L6-V2')
embedding = model.encode(text,convert_to_numpy=True)

dis = embedding.shape[1]
index = faiss.IndexFlatL2(dis)
index.add(embedding)

query =input("enter what Recepy you want")
query_vec = model.encode([query])
scores,indices = index.search(query_vec,3)

context = "\n".join([text[i] for i in indices[0]])

model = genai.GenerativeModel('models/gemini-2.5-flash')

prompt = f"""
answer the  question using only the context below and make good recepie.

context:
{context}

question:
{query}
"""
R1 = model.generate_content(prompt)
print("Answer:\n",R1.text)
```
