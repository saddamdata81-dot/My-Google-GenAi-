```python
su1 = pipeline("summarization")
context = """The Amazon rainforest is the world's largest tropical rainforest, covering ~5.5–7 million km² across nine South American nations (primarily Brazil). It is a highly biodiverse, moist broadleaf ecosystem characterized by dense canopy, high rainfall, and a critical role as a global carbon sink.
Key aspects include:
Location: Spans South America (Brazil, Peru, Colombia, etc.), covering ~40% of the continent.
Significance: Acts as a major carbon sink and regulates global climate.
Biodiversity: Home to millions of species of flora and fauna.
Habitat: Contains diverse vegetation, including tropical forests, savannas, and wetlands. """
result = su1(context,max_length = 25)
print(result)
```
